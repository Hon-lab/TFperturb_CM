## human to mouse gene conversion

```python
def extract_gene_name(name):
    parts = name.split('-')
    # Remove parts that begin with "P1P2", "P1", or "ENS"
    relevant_parts = [part for part in parts if not part.startswith(("P1P2", "P1","P2", "ENS"))]
    return '-'.join(relevant_parts) if relevant_parts else parts[0]

```

```python

# Apply the function to extract gene names from row names
TF_full_list = [extract_gene_name(name) for name in TF_full_usage.index]
```

### All the time points and phenotype

```python

filtered_df = cardiac_IMPC[cardiac_IMPC['Gene'].isin(genes_in_dict)]


# Ensure all genes in ordered_genes are in the DataFrame
df_genes = filtered_df['Gene'].tolist()
ordered_genes_in_df = [gene for gene in ordered_genes if gene in df_genes]

# Map Life Stage to numeric values (you can choose any numeric encoding, this is just an example)
life_stage_mapping = {
    'E9.5': 1,
    'E15.5': 2,
    'E18.5': 3,
    'E12.5': 4,
     'Early adult': 5,
    'Late adult': 6,
}

# Apply the mapping to the 'Life Stage' column
filtered_df['Life Stage Numeric'] = filtered_df['Life Stage'].map(life_stage_mapping)

# Create a pivot table for visualization (Phenotype as rows, Genes as columns, and Life Stage as values)
df_pivot = filtered_df.pivot_table(index='Phenotype', columns='Gene', values='Life Stage Numeric', aggfunc='first')

ordered_genes_in_df = [gene for gene in ordered_genes_in_df if gene in df_pivot.columns]

# Reorder the columns based on the filtered ordered_genes_in_df list
df_pivot = df_pivot[ordered_genes_in_df]


```

```python
heart = df_pivot[~df_pivot.index.str.contains("retina|placenta|corneal|vitelline|hemorrhage", case=False, na=False)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_life_stage_heatmap(df, selected_time_points, selected_phenotypes=None):
    """
    Generates a heatmap for selected time points and phenotypes with only colors (no numbers).
    
    Parameters:
    - df (DataFrame): The input DataFrame (pivoted form, Phenotypes as rows, Genes as columns).
    - selected_time_points (list): List of life stages to include (e.g., ['E9.5', 'E15.5', 'E18.5']).
    - selected_phenotypes (list, optional): List of phenotypes to include. If None, include all.
    """
    # Define full mapping of life stages to numeric values
    life_stage_mapping = {
        'E9.5': 1,
        'E12.5': 2,
        'E15.5': 3,
        'E18.5': 4,
        'Early adult': 5,
        'Late adult': 6
    }

    # Filter mapping for selected time points
    filtered_mapping = {k: v for k, v in life_stage_mapping.items() if k in selected_time_points}
    
    # Define color palette for selected time points
    palette = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
    cmap = sns.color_palette(palette[:len(filtered_mapping)])  # Use only needed colors

    # Copy DataFrame and filter for selected time points
    filtered_df = df.copy()
    filtered_df[~filtered_df.isin(filtered_mapping.values())] = np.nan  # Set non-selected values to NaN

    # If selected_phenotypes is provided, filter the rows
    if selected_phenotypes:
        filtered_df = filtered_df.loc[selected_phenotypes]

    # Adjust the color mapping for the heatmap
    norm_values = list(filtered_mapping.values())  # Get the values to normalize the color scale

    # Plot the heatmap (NO NUMBERS inside the cells)
    plt.figure(figsize=(12, 6))
    ax = sns.heatmap(filtered_df, cmap=cmap, linewidths=0.5, linecolor='gray', annot=False, 
                     cbar=True, vmin=min(norm_values), vmax=max(norm_values))

    # Fix the colorbar legend to show only the selected time points
    colorbar = ax.collections[0].colorbar
    colorbar.set_ticks(norm_values)  # Set tick positions
    colorbar.set_ticklabels(list(filtered_mapping.keys()))  # Set tick labels (life stages)

    # Label axes
    plt.xlabel("Genes", fontsize=12)
    plt.ylabel("Phenotypes", fontsize=12)
    plt.title("Life Stage Heatmap", fontsize=14)
    plt.savefig("/project/GCRB/Hon_lab/s438217/cNMF/TF_full_cNMF_figure_2_plots/mouse_heart_phenotype_public_dataset_early_time_points.pdf")
    # Show plot
    plt.show()
```

```python
plot_life_stage_heatmap(heart, ['E9.5', 'E12.5', 'E15.5', 'E18.5'])
```
